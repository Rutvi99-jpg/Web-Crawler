```python
"""
Libraries  Used

1. default_timer - used to monitor the performance of the program
2. bs4 - used for extracting data out of HTML and XML files
3. requests -  indicates getting or retrieving data from a specified resource
4. json - used for storing and exchanging data

PseudoCode
1. Crawl url: https://knowthychoice.in/blog - via function -> crawler
   1.1 Extract HTML page
   1.2 Find subpage
   1.3 Find cards
       1.3.1 Extract title
       1.3.2 Extract subpage url
       1.3.3 Call get_concepts with subpage_url
       1.3.4 After getting concepts add to object and return it

2. get_concepts
   2.1 Extract HTML page
   2.2 Extract Title
   2.3 Extract all the concepts
   2.4 Append all concepts in an array
    

"""


#Importing required libraries
from timeit import default_timer
import bs4
import requests
import json

#Defining crawler function
def crawler(blogpage):
    response = requests.get(blogpage)


    soup = bs4.BeautifulSoup(response.content, 'html5lib')

    data = {}

    section = soup.find('section', attrs = {'class':'blog'})
    
    #Initializing loop for extracting topic_names and navigating through links 
    for article in section.findAll('article', attrs = {'class':'blog-post-item'}):
        if article is not None:
            for row in article.findAll('div', attrs = {'class':'blog-post-content'}):
                # print(row)
                topic_name = row.h2.text.strip()
                link = row.h2.a['href']
                concepts = get_concepts(link)
                data[topic_name] = concepts

    return data

#Defining function for get_concepts
def get_concepts(topic_link):
    response = requests.get(topic_link)
    soup = bs4.BeautifulSoup(response.content, 'html5lib')

    concepts = []

    section = soup.find('section', attrs = {'class':'blog-single'})

    article = section.find('article', attrs = {'class':'blog-single-post'})
    heading_tags = ["p", "ul"]
    flag = False

    #Initializing loop for extracting concepts
    for tag in article.findAll(heading_tags):
        if tag.name == 'p' and 'Concepts Covered' in tag.text:
            flag = True
            continue

        if flag and tag.name == 'ul':
            for concept_name in tag.findAll('li'):
                concepts.append(concept_name.text)
            break
    
    return concepts

#Driver code
if __name__ == "__main__":
    START = default_timer()
    data = crawler("https://knowthychoice.in/blog/")
    print(json.dumps(data, indent=2))
    END = default_timer()
    print("Elapsed time (seconds) = " + str(END - START))
```
